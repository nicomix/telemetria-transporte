```python
import pandas as pd
```

```python
filepath = "../telemetria.csv"
data_raw = pd.read_csv(filepath)
```

```python
data_raw.head()
```

```python
# Analize nulls
print(data_raw.isna().sum())
```

```python
# Print all data types
print(data_raw.dtypes)
```

```python
# Analize duplicates
# Shows all copies of duplicate rows
duplicate_rows = data_raw[data_raw.duplicated(subset=['tracking_id', 'vehicle_id', 'event_time', 'event_type'], keep=False)]
duplicate_rows.head()
```

```python
len(duplicate_rows)
```

```python
# Analize events out of range (event_time > received_at)
outside_time = data_raw[data_raw['event_time'] > data_raw['received_at']]
outside_time.head()
```

```python
len(outside_time)
```

```python
# Analyze invalid coordinates
invalid_coordinates = data_raw[
    (data_raw['latitude'] < -90) | (data_raw['latitude'] > 90) |
    (data_raw['longitude'] < -180) | (data_raw['longitude'] > 180)
]

invalid_coordinates.head()
```

```python
len(invalid_coordinates)
```

```python
# Analyze imposible speed
imposible_speed = data_raw[
    (data_raw['speed'] < 0) | (data_raw['speed'] > 150)
]

imposible_speed.head()
```

```python
len(imposible_speed)
```
